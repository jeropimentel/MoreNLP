```python
columns=[]
for i in range(1,58):
  columns.append(i)
columns.append("Spam") #  spam = 1, no-spam = 0

data = pd.read_csv('spamdetector.txt', names=columns)
data.head()
```

```python
y = data[['Spam']]

del data['Spam']
X = data
```

```python
print(X.head())
print(y.head())
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.9, random_state=42)
```

```python

#--------INSTANTIATE-----------
nbm = MultinomialNB()
bnb = BernoulliNB()
ada = AdaBoostClassifier()
log = LogisticRegression()
#-------train the model---------
nbm.fit(X_train, y_train)
ada.fit(X_train, y_train)
bnb.fit(X_train, y_train)
log.fit(X_train, y_train)

#----------predictions------
y_pred = nbm.predict(X_test)
y_pred = ada.predict(X_test)
y_pred = bnb.predict(X_test)
y_pred = log.predict(X_test)

#----------score------------
nbm_score = nbm.score(X_test, y_test)
ada_score = ada.score(X_test, y_test)
bnb_score = bnb.score(X_test, y_test)
log_score = bnb.score(X_test, y_test)

#----------------------------

print(f" NBM score : {nbm_score:.2f} , AdaBoost score: {ada_score:.2f}, Bernoulli: {bnb_score:.2f}, Log: {log_score:.2f}")
```
